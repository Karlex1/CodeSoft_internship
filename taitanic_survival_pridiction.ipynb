```python
#survival
survival_count=dataset.Survived.value_counts()
survival_rate=100*survival_count/dataset.shape[0]
survival_rate
```

```python
survival_data=pd.concat((survival_count,survival_rate),axis=1).reset_index()
survival_data.columns=['Survived','Count','Percentage']
survival_data
```

```python
plt.figure(figsize=(3,3))
sns.barplot(survival_data,x=('Not-Survived','Survived'),y='Percentage',width=0.50)
plt.xlabel('Survival status')
plt.ylabel('Percentage')
plt.title('Figure to show Survival_data')
```
